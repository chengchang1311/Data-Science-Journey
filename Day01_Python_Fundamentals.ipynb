```python
patients = [
    {"name": "Nguyen Van An",  "age": 45, "glucose": 180, "bmi": 28.5},
    {"name": "Tran Thi Bich",  "age": 32, "glucose": 95,  "bmi": 21.0},
    {"name": "Le Van Cuong",   "age": 58, "glucose": 220, "bmi": 31.2},
    {"name": "Pham Thi Dung",  "age": 27, "glucose": 105, "bmi": 17.8},
    {"name": "Hoang Van Em",   "age": 61, "glucose": 148, "bmi": 26.4},
]

high_risk_count = 0
for i, patient in enumerate(patients):
    # Bước 1: Lấy dữ liệu từ dict ra
    ten     = patient["name"]
    tuoi    = patient["age"]
    glucose = patient["glucose"]
    bmi     = patient["bmi"]

    # Bước 2: Phân loại — em tự viết điều kiện vào đây
    if glucose >=126 or bmi >=30:
        phanloai = "⚠️ NGUY CƠ CAO"
        high_risk_count += 1
    else:
        phanloai = "✅ Bình thường"
     # Bước 3: In thông tin — em tự viết f-string vào đây
    print(f"Bệnh nhân{i + 1}: Glucose{glucose} -> {phanloai}")

# Bước 4: In tổng — nằm NGOÀI vòng lặp
print(f"\nTổng nguy cơ cao: {high_risk_count}/{len(patients)}")
```
